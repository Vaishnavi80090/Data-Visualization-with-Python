# Basic pyplot function

```python
x_int=np.arange(1,21)
y_int=np.random.randint(1000,size=20)
```

```python
plt.scatter(x_int,y_int,marker='*',c='g')
plt.xlabel('x_int')
plt.ylabel('y_int')
plt.title('random chart for practice',fontsize=15,fontstyle='italic',c='r')
plt.xticks(x_int,rotation=-5)
plt.savefig('first_matplotlib_plot.png')
plt.grid()
```

```python
plt.scatter(x_int,y_int,marker='*',c='g',label='stars')
x_triangle=np.arange(1,21)
y_triangle=np.random.randint(1000,size=20)
plt.scatter(x_triangle,y_triangle,marker="v",c='r',label='triangle')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('random chart for practice ',fontsize=14,fontstyle='italic',c='orange')
plt.xticks(x_int,rotation=-5)
plt.legend()
```

# Bar Chart

```python
days=['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
sales=np.random.randint(50000,size=7)
plt.bar(days,sales,color=['r','b','g','maroon','orange','y','m'])
plt.xticks(days,rotation=90)
plt.xlabel('days',fontsize=14,fontstyle='italic',c='maroon')
plt.ylabel('sales',fontsize=14,fontstyle='italic',c='maroon')
plt.title('random chart of bar',fontsize=15,fontstyle='italic',backgroundcolor='yellow',c='maroon')
```

# Variation in Bar chart | Stocked Bar Chart

```python
days=['sunday','monday','tuesday','wednesday','thursday','friday','saturday']
shop_sales1=np.random.randint(50000,size=7)
shop_sales2=np.random.randint(50000,size=7)
plt.xticks(rotation=90)
plt.bar(days,shop_sales1,label='shop1',color='orange')
plt.bar(days,shop_sales2,label='shop2',color='r')
plt.legend()
```

# Variations in bar chart | Doged chart

```python
x_indices=np.arange(0,len(days))
x_indices_shop2=np.arange(0,len(days))

width=0.3
x_indices_shop2=x_indices_shop2+width
plt.bar(x_indices,shop_sales1,width,label='shop1',color='g')
plt.bar(x_indices_shop2,shop_sales2,width,label='shop2',color='y')
plt.legend()
plt.xticks((x_indices+x_indices_shop2)/2,days)
plt.xlabel(('days'),color='maroon')
plt.ylabel(("sales price"),color='maroon')
plt.title("per day sales",color='maroon',backgroundcolor='y')
plt.xticks(rotation=90)
```

# Pie chart

```python
segment='AWS','Amazon','Prime','Other'
revenue=np.random.randint(20000,size=4)
plt.pie(revenue,labels=segment,autopct='%1.1f%%',shadow=True)
plt.title('Amazon revenue split by category',fontsize=15,fontstyle='italic',c='g')

```

# Heat Map

```python
data=pd.DataFrame(np.random.rand(1000,2),columns=list('AB'))
```

```python
plt.hist2d(data['A'],data['B'],bins=40)
plt.colorbar()
plt.title('heatmap of frequency of values')
```

There is another function to make a heatmap via Pyplot - imshow() Let's make a heatmap using the imshow function.
The drawback here is that imshow doesn't work with Panda data frames, and adjusting the bins is a bit complicated. However, it 
can be used where data to be visualized is not huge.


```python
data_imshow=np.random.rand(10,10)
plt.imshow(data_imshow,cmap='hot')
plt.colorbar()
```

# Subplot

These are extremely useful to show multiple charts in the same frame, especially if those are interlinked and a story can be told by looking at a picture collectively.



```python
plt.subplot(2,2,1)
plt.plot(x_int,y_int,'red')
plt.subplot(2,2,2)
plt.plot(x_int,y_int,'g')
plt.subplot(2,2,3)
plt.plot(x_int,y_int,'bo')
plt.subplot(2,2,4)
plt.plot(x_int,y_int,'v')
```
