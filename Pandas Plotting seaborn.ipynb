##### Inferences â€“

## KDE Plot

KDE stands for Kernel Density Estimate.
It plots an estimate of the Probability distribution sample of a variable. This graph has great application in statistics and feature
engineering.

```python
iris.plot.kde()
```

Inferences -
â— Sepal length takes the highest values compared to the four variables, meaning Sepal length is longest by absolute value in general.
The opposite is true for Petal Width.
â— Most plants have more or less similar Sepal Width (Close to its mean)
â— Petal length's range is maximum, which can be a major determining factor in the identity of a flower from another.

## Distplot


```python
sns.distplot(iris['sepal_width'],hist_kws={'color':'teal','edgecolor':'black'},kde_kws={'color':'Navy'},bins=20)
```

```python
sns.histplot(iris['sepal_width'],bins=20,kde=True)
```

### jointplot

```python
sns.jointplot(iris['sepal_length'],iris['sepal_width'],alpha=0.7,color='teal')
```

```python
sns.jointplot(iris['sepal_length'],iris['sepal_width'],alpha=0.7,color='teal',kind='hex')
```

Inferences â€“
â— Mean of the Sepal length and Sepal width lie â€“ around 6 and 3, respectively.
â— There are 2 clusters of areas where density is more than another part:


```python
sns.jointplot(iris['sepal_length'],iris['sepal_width'],kind='reg')
```

```python
sns.set_style('white')
sns.jointplot(iris['sepal_length'],iris['sepal_width'],kind='kde',space=0)
```

### Pairplot

```python
sns.pairplot(iris,hue='species',diag_kind='kde',size=2)
```

### Swarmplot and Violinplot


```python
plt.subplot(121)
sns.swarmplot('species','petal_length',data=iris)
plt.subplot(122)
sns.violinplot('species','petal_length',data=iris)
```

Inferences -
â— Setosa has the smallest Petal length of all flowers, while Virginica has the longest.
â— Setosa also has a sharper curve, meaning most of the flowers had lengths close to the mean. The same is not true for the other two
flowers having a flatter distribution on Petal length

### countplot

### barplot

```python
sns.barplot(x='petal_width',y='species',data=iris)
```

Inferences â€“
1. Setosa has the smallest Petal width, Virginica is the largest.
2. The error bars on top represent that Setosa also has a lower deviation from the mean than the other two flowers for Petal Width.
